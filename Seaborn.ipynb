```python
# for distribution plots
import seaborn as sns
```

### Distribution plots
##### distplot
##### pairplot
##### jointplot

```python
df=sns.load_dataset("tips")
```

```python
df.corr()
```

```python
#correlation with heatmap
sns.heatmap(df.corr())
```

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') # kind = 'reg' is regression, the other plots depict the distribution
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.pairplot(df) # histogram along the digonal gives the distribution of the variable, scatterplot along upper and lower
# triangle gives the plot of the two variables
```

```python
sns.pairplot(df,hue='sex') # hue can be any categorical variable
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

### Categorical plots
####

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
# Box plot
# A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.
```

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df)
```

```python
#Violin Plot
#Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot
sns.violinplot(x="total_bill", y="day",hue = 'smoker' ,data=df)
```
